```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import re
#import emoji
import demoji
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

def remove_stopwords(input_string):
    stop_words = set(stopwords.words('spanish'))  # change 'english' to your language if needed
    word_tokens = word_tokenize(input_string)
    filtered_sentence = [word for word in word_tokens if not word in stop_words]
    return ' '.join(filtered_sentence)  # join the words back into a string

def remove_punctuation(input_string):
    # Make a translator object that will replace punctuation with None
    translator = str.maketrans('', '', string.punctuation)
    # Use the translator
    no_punct = input_string.translate(translator)
    return no_punct

# Download the latest emoji codes.
demoji.download_codes()

def remove_emojis(text):
    # Use the sub() method to replace all found emojis with an empty string
    no_emoji_text = demoji.replace(text, "")
    # remove breaks
    text_clean = no_emoji_text.replace('\n', ' ')

    return text_clean

#tweet = "Hello World 🌍 😊"
#clean_tweet = remove_emojis(tweet)
#print(clean_tweet)  # Output: Hello World 


def scrape_tweets_user(username, num_tweets):
    # Initialize a list to hold all the tweepy Tweets
    tweets_list1 = []

    # Using TwitterSearchScraper to scrape data and append tweets to list
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f'from:{username}').get_items()):
        if i > num_tweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.hashtags, tweet.user.username, tweet.likeCount, 
                             tweet.retweetCount, tweet.rawContent, tweet.retweetedTweet, tweet.mentionedUsers, tweet.url])

    # Creating a dataframe from the tweets list above 
    tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'TweetId', 'Hashtags', 'Username', 'LikeCount', 
                                                      'ReTweetCount', 'RawContent', 'Rettweeted', 'MentionedUsers', 'Url'])
    
    return tweets_df1

#tweets_df = scrape_tweets('bbva_espana', 1000)
#print(tweets_df)

def scrape_tweets_hashtag(hashtag, num_tweets):
    # Initialize a list to hold all the tweepy Tweets
    tweets_list1 = []

    # Using TwitterSearchScraper to scrape data and append tweets to list
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f'#{hashtag}').get_items()):
        if i > num_tweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.hashtags, tweet.user.username, tweet.likeCount, 
                             tweet.retweetCount, tweet.rawContent, tweet.retweetedTweet, tweet.mentionedUsers, tweet.url])

    # Creating a dataframe from the tweets list above 
    tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'TweetId', 'Hashtags', 'Username', 'LikeCount', 
                                                      'ReTweetCount', 'RawContent', 'Rettweeted', 'MentionedUsers', 'Url'])
    return tweets_df1

def scrape_tweets_mention(username, num_tweets):
    # Initialize a list to hold all the tweepy Tweets
    tweets_list1 = []

    # Using TwitterSearchScraper to scrape data and append tweets to list
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f'@{username}').get_items()):
        if i > num_tweets:
            break
        tweets_list1.append([tweet.date, tweet.id, tweet.hashtags, tweet.user.username, tweet.likeCount, 
                             tweet.retweetCount, tweet.rawContent, tweet.retweetedTweet, tweet.mentionedUsers, tweet.url])

    # Creating a dataframe from the tweets list above 
    tweets_df1 = pd.DataFrame(tweets_list1, columns=['Datetime', 'TweetId', 'Hashtags', 'Username', 'LikeCount', 
                                                      'ReTweetCount', 'RawContent', 'Rettweeted', 'MentionedUsers', 'Url'])
    return tweets_df1
```

```python
#mentions
tweets_mention_BBVA = scrape_tweets_mention('bbva_espana', 1000)
tweets_mention_BBVA['CleanContent'] = tweets_mention_BBVA['RawContent'].str.lower()
tweets_mention_BBVA['Clean'] = tweets_mention_BBVA['CleanContent'].apply(remove_punctuation)
#tweets_mention_BBVA['Final'] = tweets_mention_BBVA['CleanContent'].apply(remove_stopwords)
tweets_mention_BBVA['Bank'] = 'BBVA'

tweets_mention_BS = scrape_tweets_mention('santander_es', 1000)
tweets_mention_BS['CleanContent'] = tweets_mention_BS['RawContent'].str.lower()
tweets_mention_BS['Clean'] = tweets_mention_BS['CleanContent'].apply(remove_punctuation)
#tweets_mention_BS['Final'] = tweets_mention_BS['CleanContent'].apply(remove_stopwords)
tweets_mention_BS['Bank'] = 'BancoSantander'

tweets_mention_BA = scrape_tweets_mention('bancosantander', 1000)
tweets_mention_BA['CleanContent'] = tweets_mention_BA['RawContent'].str.lower()
tweets_mention_BA['Clean'] = tweets_mention_BA['CleanContent'].apply(remove_punctuation)
#tweets_mention_BA['Final'] = tweets_mention_BA['CleanContent'].apply(remove_stopwords)
tweets_mention_BA['Bank'] = 'BancoSantander'

tweets_mention_CB = scrape_tweets_mention('caixabank', 1000)
tweets_mention_CB['CleanContent'] = tweets_mention_CB['RawContent'].str.lower()
tweets_mention_CB['Clean'] = tweets_mention_CB['CleanContent'].apply(remove_punctuation)
#tweets_mention_CB['Final'] = tweets_mention_CB['CleanContent'].apply(remove_stopwords)
tweets_mention_CB['Bank'] = 'CaixaBank'

#published
tweets_pub_BBVA = scrape_tweets_user('bbva_espana', 1000)
tweets_pub_BBVA['CleanContent'] = tweets_pub_BBVA['RawContent'].str.lower()
tweets_pub_BBVA['Clean'] = tweets_pub_BBVA['CleanContent'].apply(remove_punctuation)
tweets_pub_BBVA['Bank'] = 'BBVA'

tweets_pub_BS = scrape_tweets_user('santander_es', 1000)
tweets_pub_BS['CleanContent'] = tweets_pub_BS['RawContent'].str.lower()
tweets_pub_BS['Clean'] = tweets_pub_BS['CleanContent'].apply(remove_punctuation)
tweets_pub_BS['Bank'] = 'BancoSantander'

tweets_pub_CB = scrape_tweets_user('caixabank', 1000)
tweets_pub_CB['CleanContent'] = tweets_pub_CB['RawContent'].str.lower()
tweets_pub_CB['Clean'] = tweets_pub_CB['CleanContent'].apply(remove_punctuation)
tweets_pub_CB['Bank'] = 'CaixaBank'
```

```python
# Append DF
dfs_mention = [tweets_mention_BBVA, tweets_mention_BS, tweets_mention_CB,tweets_mention_BA] 
tweets_df_mention = pd.concat(dfs_mention)

dfs_banca = [tweets_pub_BBVA, tweets_pub_BS, tweets_pub_CB] 
tweets_df_pub = pd.concat(dfs_banca)
```

```python
tweets_df_mention.to_csv('BanksSpainMention.csv', index=False, quotechar = '"')
tweets_df_pub.to_csv('BanksSpainPub.csv', index=False, quotechar = '"')
```
